```python
X_train = X_train / 255
X_test = X_test / 255
```

## Using ANN for classification

```python
from keras.layers import Input, Flatten, Dense
from keras.models import Model
```

```python
from keras import Sequential
model = Sequential()
```

```python
input = Input(shape=(28,28))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(128, activation='relu'))
model.add(Dense(32,activation='relu'))
```

```python
model.add(Dense(10, activation='sigmoid'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs = 25, validation_split = 0.2)
```

```python
model.summary()
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend()
```

```python
model.evaluate(X_test,y_test)
```

Thus, we have accuracy of the model around 97%.

```python
X_train = X_train.reshape(-1,28,28,1)
X_train.shape
```

## Using CNN for classification

```python
from keras import layers, models

input_layer = layers.Input(shape=(28, 28, 1))

model = models.Sequential([input_layer,
    layers.Conv2D(30, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(100, activation='relu'),
    layers.Dense(10, activation='sigmoid')
])

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=5)
```

```python
y_train[:5]
```

```python
model.evaluate(X_test,y_test)
```

Thus, with CNN we get an impressive overall accuracy above 98%, which is more than that predicted with ANN.
